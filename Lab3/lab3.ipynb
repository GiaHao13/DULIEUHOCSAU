```python
#Vấn đề 1: Tiến hành tải dữ liệu vào chương trình ứng dụng Python và giải
column_names=["ID", "Name", "Age", "Weight", "m0006", "m0612","m1218", "f006", "f0612", "f1218"]
patientheartrate = pd.read_csv("patient_heart_rate.csv", names = column_names)
print(patientheartrate.head())
```

```python
#Vấn đề 2: Xử lý vấn đề một cột lưu hỗn hợp nhiều dữ liệu, ở đây là cột ―Name‖ chứa bao gồm ―Firstname‖ và ―Lastname‖, giải pháp là ta sẽ tách ra làm 2 cột
patientheartrate[['Firstname','Lastname']] = patientheartrate['Name'].str.split(expand=True)
patientheartrate = patientheartrate.drop('Name', axis=1)
print (patientheartrate)
```

```python
#Vấn đề 3: Cột Weight có vấn đề về không thống nhất các đơn vị đo lường trong dữ liệu. Ta sẽ chuyển các đơn vị về thành đơn vị chuẩn ―kg‖
weight=patientheartrate['Weight']

for i in range(0,len(weight)):
  x = str(weight[i])
  if "lbs" in x[-3:]:
    #remove the lbs from the value
    x = x[:-3:]
    float_x = float(x)
    y = int(float_x/2.2)
    y = str(y)+"kgs"
    weight[i] = y
print (patientheartrate)
```

```python
#Vấn đề 4: Vấn đề về xuất hiện dòng dữ liệu rỗng (không có giá trị: NaN). Giải pháp có thể đưa ra là xóa bỏ
patientheartrate.dropna(how="all", inplace=True)
print(patientheartrate)
```

```python
#Vấn đề 5: Có nhiều dòng dữ liệu bị trùng lắp thông tin hoàn toàn[fullname, lastname, age, weight,....], giải pháp đưa ra là chỉ giữ lại một dòng dữ liệu, tuy nhiên giải pháp phải dựa trên nghiệp vụ của tập dữ liệu và quan sát của người xử lý.
patientheartrate = patientheartrate.drop_duplicates(subset=['Firstname', 'Lastname', 'Age', 'Weight'])
print (patientheartrate)
```

```python
#Vấn đề 6: Xuất hiện dữ liệu bị ảnh hưởng bởi lỗi non-ASCII, không định dạng ASCII. Giải pháp: Tùy vào nghiệp vụ ta có thể: xóa dữ liệu tại đó, thay thế bằng dữ liệu khác hoặc thay bằng việc đánh dấu bằng một kí tự khác (ví dụ: ‗warning‘)
patientheartrate.Firstname.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
patientheartrate.Lastname.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
print(patientheartrate)
```
