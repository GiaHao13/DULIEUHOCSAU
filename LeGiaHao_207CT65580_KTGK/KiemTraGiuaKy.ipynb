```python
import pandas as pd
import numpy as np
```

```python
# Cau 1: 
data = {
    'Name': ['Alice', 'Bob', 'Charlie','David','Eva','frank','Grace','Hannah','Ivan','Jack','Kelly','Liam','Mona','Nina','Oscar'],
    'Age': [25,30,35,28,22,45,34,31,27,29,33,40,26,32,36],  
    'Salary': [50000,60000,70000,55000,52000,80000,72000,68000,61000,59000,63000,77000,53000,66000,75000]  
}

df = pd.DataFrame(data)

print(df)
print(df.columns)
print(df.index)
```

```python
# Cau2: 
df.head(15)
```

```python
# Cau 3: 
age_df = df[df['Age'] > 28]
print(age_df)
```

```python
# Cau 4:
average_salary = df['Salary'].mean()
print(average_salary)
```

```python
# Cau 5:
group_df = df.groupby('Age')['Salary'].sum().reset_index()
print(group_df)
```

```python
# Cau 6:
decrease_df = df.sort_values (by = 'Salary', ascending = False)
print(decrease_df)
```

```python
# Cau 7:
import matplotlib.pyplot as plt

plt.bar(df['Name'], df['Age'], color = 'green')
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age')
plt.show()
```

```python
# Cau 8:
plt.plot(df['Name'], df['Salary'], color = 'green')
plt.title('Salary')
plt.xlabel('Name')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
# Cau 9:
plt.pie(df['Age'], labels=df['Name'])
plt.title('Age')
plt.show()
```

```python
# Cau 10:
import seaborn as sns

bieudophantan =df[['Age', 'Salary']]
sns.pairplot(bieudophantan)
```

```python
# Cau 11:
test = df.isna().sum()
print(test)
```

```python
# Cau 12:
average_age = df['Age'].mean()
df.loc[df['Age'] > 30, 'Age'] = average_age
print(df)
```

```python
# Cau 13:
df['Age_normalized'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
print(df)
```

```python
# Cau 14:
def sapxep_age(age):
    if age <= 30:
        return 'young'
    elif 30 < age < 60:
        return 'middle_aged'
    else:
        return 'old'
df['age_group'] = df['Age'].apply(sapxep_age)
print(df)
```

```python
# Cau 15:
df['percentage'] = df['Salary'].pct_change() * 100
print(df)
```

```python
# Cau 16:
df.drop_duplicates(subset=['Name', 'Age', 'Salary'])
print(df)
```

```python
# Cau 17:
df.to_csv('baikiemtraso1.csv', index=True)
```
